```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
resp = requests.get('https://ebird.org/region/IN-PB-SH?yr=all')
soup = BeautifulSoup(resp.text, 'html.parser')
```

```python
df = pd.DataFrame(columns = ['Species', 'eBird Link', 'Info'])
```

```python
for i, species in enumerate(soup.find_all('div', 'Observation-species')):
    try:
        df.loc[i, 'Species'] = str(species.a.span.string)
        df.loc[i, 'eBird Link'] = species.a['href']
    except:
        break
```

```python
df
```

```python
df.drop(0, inplace = True)
```

```python
done_indices = df[df['Info'].notna()].index.tolist()

for i, link in enumerate(df['eBird Link']):
    if i+1 in done_indices: continue
    try:
        birdResp = requests.get('https://ebird.org'+link)
    except:
        time.sleep(10)
        birdResp = requests.get('https://ebird.org'+link)
    birdSoup = BeautifulSoup(birdResp.text, 'html.parser')
    df.loc[i+1, 'Info'] = str(birdSoup.find('p', 'u-stack-sm').string)
```

```python
df
```

```python
df.to_csv('Birds Data.csv')
```
